### Function to Calculate Centroids of the rectangular bounding boxes
### Classification of Myeloid subtypes

```python
def X_centroid(row):
    X = int(row['XMin'] + row['XMax'])/2

    return int(X)

def Y_centroid(row):
    Y = int(row['YMin'] + row['YMax'])/2
    return int(Y)

### Function of myeloid cell Classification


def classification(row):
    
    if (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] == 0) & (row['Opal_620_Classification'] >=1) & (row['Opal_690_Classification'] == 0):
        return 'CM0'
    elif (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] == 0) & (row['Opal_620_Classification'] >=1) & (row['Opal_690_Classification'] >=1):
        return 'CM1'
    elif (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] >= 1) & (row['Opal_620_Classification'] ==0) & (row['Opal_690_Classification'] ==0):
        return 'CM4'
    elif (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] == 0) & (row['Opal_620_Classification'] ==0) & (row['Opal_690_Classification'] == 0):
        return 'CM_other'
    elif (row['Opal_520_Classification'] ==0) & (row['Opal_570_Classification'] >= 1) & (row['Opal_620_Classification'] ==0) & (row['Opal_690_Classification'] == 0):
        return 'RNASE'
    elif (row['Opal_520_Classification'] ==0) & (row['Opal_570_Classification'] == 0) & (row['Opal_620_Classification'] >=1) & (row['Opal_690_Classification'] == 0):
        return 'CD300'
    elif (row['Opal_520_Classification'] ==0) & (row['Opal_570_Classification'] == 0) & (row['Opal_620_Classification'] == 0) & (row['Opal_690_Classification'] >=1):
        return 'CD36'
    elif (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] == 0) & (row['Opal_620_Classification'] == 0) & (row['Opal_690_Classification'] >=1):
        return 'CM_CD36'
    elif (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] >=1) & (row['Opal_620_Classification'] >=1) & (row['Opal_690_Classification'] ==0):
        return 'CM_RNASE_CD300'
    elif (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] >=1) & (row['Opal_620_Classification'] ==0) & (row['Opal_690_Classification'] >=1):
        return 'CM_RNASE_CD36'
    elif (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] >=1) & (row['Opal_620_Classification'] >=1) & (row['Opal_690_Classification'] >=1):
        return 'CM_RNASE_CD300_CD36'
    else:
        return 'NC'
    

def subtypes(row):
    if (row['Classification'] == "CM0"):
        return "CM0"
    elif (row['Classification'] == "CM1"):
        return "CM1"
    elif (row['Classification'] == "CM4"):
        return "CM4"
    else:
        return "others"


def calculate_centroid_classification():
    
    prf = []

    for i, p in enumerate(data['Image_ID'].unique()):
        
        dt = data.loc[data['Image_ID'] == str(p)]
    
    # Use the height and width to calculate the area

        dt['X'] = dt.apply(X_centroid, axis=1)
        dt['Y'] = dt.apply(Y_centroid, axis=1)
        
        dt['Classification'] = dt.apply(classification, axis=1)
        dt['Classification_1'] = dt.apply(subtypes, axis=1)
    
        prf.append(dt)

    data_new = pd.concat(prf)
      
    
    return data_new
    
data = calculate_centroid_classification() 




```

## Creating Mask 

### Note:
1) Annotated regions are filled using polygon function(Skimage)

2) Cell coordinates were matched with masked region and separate dataframe is generated

#### Glomerulus Mask

```python
from skimage.draw import polygon


def Intersection(dtt, mask): 
    return set(dtt).intersection(set(mask))



gf = []

for i, p in enumerate(data.Image_ID.unique()):
    
    anno = d.loc[d['Image_ID'] == str(p)]
    dt= data.loc[data['Image_ID'] == str(p)]
    
    glom_id = anno.loc[anno['Annotation'] == "gloms"]['Region_ID'].unique().tolist()
    
    if len(glom_id) > 0:
        
          
        
        for gl in glom_id:
            
            poly = anno[(anno.Annotation == 'gloms') & (anno.Region_ID == int(gl))]
            r, c = polygon(np.array(poly.X),
                         np.array(poly.Y))
            mask = [m for m in zip(r, c)]
    
            dtt = [t for t in zip(dt.X, dt.Y)]
            d_mask = Intersection(dtt,mask)
            g = pd.DataFrame(d_mask, columns=['X', 'Y'])
            g['mask'] = 'gloms'
            g['Region_ID'] = str(gl)
            g['Image_ID'] = str(p)
        
            gf.append(g)
        
gmask = pd.concat(gf)


gmask.head()


```

#### Vessel Mask

```python
vf =[]

for i, p in enumerate(d.Image_ID.unique()):
    anno = d.loc[d['Image_ID'] == str(p)]
    dt= data.loc[data['Image_ID'] == str(p)]
    
    vessel_id = anno.loc[anno['Annotation'] == "vessels"]['Region_ID'].unique().tolist()
    
    if len(vessel_id) > 0:
        
        for ve in vessel_id:
            
            
            poly = anno[(anno.Annotation == 'vessels') & (anno.Region_ID == ve)]
            r, c = polygon(np.array(poly.X),
                         np.array(poly.Y))
            mask = [m for m in zip(r, c)]
    
            dtt = [t for t in zip(dt.X, dt.Y)]
            d_mask = Intersection(dtt,mask)
            v = pd.DataFrame(d_mask, columns=['X', 'Y'])
            v['mask'] = 'vessels'
            v['Region_ID'] = str(ve)
            v['Image_ID'] = str(p)
        
            vf.append(v)
           
        
vmask = pd.concat(vf)
vmask


fmask = pd.DataFrame(gmask.append(vmask)) 

fmask['mask'].unique()

fmask.head()
```

## Distance Mapping

### Spatial Distance function and calculation of other metrics

```python
import math

def spatialdistances(mask):
  
    gp= []

    mask_list = ["gloms", 'vessels']

    for i, imgid  in enumerate(final['Image_ID'].unique()):
    
        df = final.loc[(final['Image_ID'] == imgid) & (final['mask'] == mask)]
        anno = d.loc[(d['Image_ID'] == imgid) & (d['Annotation'] == mask)]
        
        if len(df) > 0:
            for rid in sorted(df.Region_ID.unique()):
                af = anno.loc[anno['Region_ID'] == rid]
                gf = df.loc[df['Region_ID'] == rid]       
        
                dist= []
                coordinates = []
        
                for i in zip(af.X, af.Y):
                
                    for j in zip(gf.X, gf.Y):
                    
                        cor = [i,j]
                        dt = math.sqrt((i[0] - j[0])**2 + ((i[1] - j[1])**2)) * 0.495952562137432
                        dist.append(dt)
                        coordinates.append(cor)
    
                    distance = pd.DataFrame(dist, columns=['Distance']) 
                    coor = pd.DataFrame(coordinates, columns=['cord_1', 'cord_2'])   
                    pf = pd.concat([coor, distance], axis =1)
            
                lst = []
            
                for cor in pf.cord_2.unique():          
                    tm = pf.loc[pf['cord_2'] == cor]
                    tm2 = tm.loc[tm['Distance'] == min(tm['Distance'])]
                    lst.append(tm2)
                vtt = pd.concat(lst) 
            
            #vtt = vt.drop_duplicates(subset=["cord_1","cord_2", "Distance"])
                vtt['Region_ID'] = str(rid)
                vtt['mask'] = mask
                vtt['Image_ID'] = imgid
                vtt[['X1', 'Y1']] = pd.DataFrame(vtt['cord_1'].tolist(), index=vtt.index) 
                vtt[['X', 'Y']] = pd.DataFrame(vtt['cord_2'].tolist(), index=vtt.index)
                ft = vtt.loc[:, ['Image_ID', 'X', 'Y', 'Region_ID', 'mask','Distance']]
                ft['Distance_normalized'] = ft['Distance'].apply(lambda x: int(x -  np.mean(ft.Distance))/np.std(ft.Distance))
            ## Dropping columns inorder to merge with the original dataframe to avoid duplicate columns
                pmd = ft.drop(columns=['Region_ID','mask']).reset_index(drop=True)
                prf = final.loc[(final['Image_ID'] == imgid) & (final['mask'] == mask) & (final['Region_ID'] == rid)].reset_index(drop = True)
                tmp =  prf.merge(pmd,how='left', on=['X', 'Y', 'Image_ID'])
                fd = tmp.drop_duplicates(subset=["Object_Id","X", "Y"])
            
                fd['CM4'] = fd['Classification_1'].apply(lambda x: x == "CM4").sum()
                fd['CM1'] = fd['Classification_1'].apply(lambda x: x == "CM1").sum()
                fd['CM0'] = fd['Classification_1'].apply(lambda x: x == "CM0").sum()
                fd['Total_Cellcount'] = len(fd['Classification_1'])
            
                fd['CM4_prop'] = round(fd.CM4[0] / fd.Total_Cellcount[0], 4)
                fd['CM1_prop'] = round(fd.CM1[0] / fd.Total_Cellcount[0], 4)
                fd['CM0_prop'] = round(fd.CM0[0] / fd.Total_Cellcount[0], 4)
            

                fd['Average_Distance'] = round(fd['Distance'].mean(),4)
                fd['CM4_meandist'] = round(fd.loc[fd['Classification'] == "CM4"]['Distance'].mean(), 4)
                fd['CM1_meandist'] = round(fd.loc[fd['Classification'] == "CM1"]['Distance'].mean(), 4)
                fd['CM0_meandist'] = round(fd.loc[fd['Classification'] == "CM0"]['Distance'].mean(), 4)           
                gp.append(fd)
            
            pm = pd.concat(gp)
    
    
    return pm

pm = spatialdistances("gloms")
vs = spatialdistances("vessels")

gv = pd.DataFrame(pm.append(vs)) 
    
```

## Calculation of Density score

```python
def density_rid():

    scores_rid =[]

    for i, p in enumerate(gv['Image_ID'].unique()):
        
        tmp = gv.loc[gv['Image_ID'] == str(p)]
        
        for j in tmp['mask'].unique():
            
            tmp1 = tmp.loc[tmp['mask'] == str(j)]
                     
            for rid in tmp1['Region_ID'].unique():

                dgn = tmp1.loc[tmp1['Region_ID'] == rid]       
                temp1 = dgn[['X', 'Y']].to_numpy()
                kde_glom = KernelDensity(kernel='gaussian', bandwidth=1).fit(temp1)
                kscore_glom = kde_glom.score_samples(temp1)
                probX_glom = np.exp(kscore_glom)
                gx = pd.DataFrame(probX_glom, columns=['Density_gloms'])
                gx['Region_ID'] = int(rid)
                gx['mask'] = str(j)
                gx['Image_ID'] = str(p)
                scores_rid.append(gx)     
        scores_gloms = pd.concat(scores_rid)
    
    return scores_gloms

density_rid()



    
def density_subtypes():
    
    scores_cells =[]
    for i, p in enumerate(gv['Image_ID'].unique()):
                
        tmp = gv.loc[gv['Image_ID'] == str(p)]
        
        for j in tmp['mask'].unique():        
            
            tmp1 = tmp.loc[tmp['mask'] == str(j)]
                     
            for rid in tmp1['Region_ID'].unique():
        
                dgn = tmp1.loc[tmp1['Region_ID'] == rid]
                
                for k in dgn['Classification_1'].unique():
                    
                
                    ct = dgn.loc[dgn['Classification_1'] == str(k)]
                    temp1 = ct[['X', 'Y']].to_numpy()
                    kde_subtypes = KernelDensity(kernel='gaussian', bandwidth=1).fit(temp1)
                    kscore_sub = kde_subtypes.score_samples(temp1)
                    probX_sub = np.exp(kscore_sub)
                    clx = pd.DataFrame(probX_sub, columns=['Density_subtypes'])
                    clx['Classification_1'] = str(k)
                    clx['Region_ID'] = int(rid)
                    clx['mask'] = str(j)
                    clx['Image_ID'] = str(p)
                    scores_cells.append(clx)     
    sc_cells = pd.concat(scores_cells)
    
    return sc_cells
    
d1 = density_rid().reset_index(drop=True)
d2 = density_subtypes().reset_index(drop=True)



dens = pd.concat([d2, d1], axis=1)

dens.head()

density = dens.loc[:, ~dens.columns.duplicated()]
density.head()


density_columns = density.loc[:, ['Image_ID', 'Region_ID','Classification_1', 'mask', 'Density_gloms', 'Density_subtypes']].reset_index(drop=True)

density_columns.shape

density['mask'].unique()
```
