## Paths of  Data and Annotations

```python

def fileslist():
    dir_data = '../data' + '/'+ foldername 
    datacsvs = os.listdir(dir_data)
    flist = []
    for f in datacsvs:
        if f.endswith('csv'):
            fpath = os.path.join(dir_data, f)
            flist.append(fpath)
    return sorted(flist)

def annolist():
    dir_anno = '../data'+ '/'+ foldername + '/' + 'annotations'
    annocsvs = os.listdir(dir_anno)
    annolist = []
    for i in annocsvs:
        if not "job" in i:
            annopath = os.path.join(dir_anno, i)
            annolist.append(annopath)
    return sorted(annolist)


flist = sorted(fileslist())

annolist = sorted(annolist())

len(annolist)

annolist
# print(flist)

```

```python
def getfilename(f):
    f = os.path.split(f)[1]
    f = f.split('_')
    f= f[2]
    fname = f.replace('.tif', '')
    return fname



def loading_data():
    
    prf = []
    
    for i, f in enumerate(flist):
        dt = pd.read_csv(f)
        dt.columns = [col.replace(' ', '_') for col in dt.columns]
        
        dt['Image_ID'] = getfilename(f)
        
#         dt["Image_Location"] = dt["Image_Location"].apply(lambda x: x.split('\\')[-1])
#         dt["Image_Location"] = dt["Image_Location"].apply(lambda x: x.partition('.tif')[0])
        
        
#         dt= (dt
#             .rename(columns={'Image_Location': 'Image_ID'})
#             .drop(columns = ['Analysis_Region', 'Analysis_Inputs'])  
#            )
        
        columns_to_keep = ['Image_ID','Object_Id', 'XMin', 'XMax', 'YMin', 'YMax',
       'Opal_520_Copies', 'Opal_520_Area_(µm²)', 'Opal_520_Classification',
       'Opal_520_Cell_Intensity', 'Opal_520_Avg_Intensity', 'Opal_570_Copies',
       'Opal_570_Area_(µm²)', 'Opal_570_Classification',
       'Opal_570_Cell_Intensity', 'Opal_570_Avg_Intensity', 'Opal_620_Copies',
       'Opal_620_Area_(µm²)', 'Opal_620_Classification',
       'Opal_620_Cell_Intensity', 'Opal_620_Avg_Intensity', 'Opal_690_Copies',
       'Opal_690_Area_(µm²)', 'Opal_690_Classification',
       'Opal_690_Cell_Intensity', 'Opal_690_Avg_Intensity', 'Cell_Area_(µm²)',
       'Cytoplasm_Area_(µm²)', 'Nucleus_Area_(µm²)', 'Nucleus_Perimeter_(µm)',
       'Nucleus_Roundness']
        
        dm = dt.loc[:, columns_to_keep]
    
        prf.append(dm)
        
    data = pd.concat(prf)
    
    return data




def loading_annotation():
    
    prf = []
    
    for i, p in enumerate(annolist):  
        
        path, fname = os.path.split(p)
        f = fname.split('.csv')[0]
        f1 = f.split('_')
        newname = f1[2]
        
        
        anno = pd.read_csv(p, skiprows=[0])
        
        columns_to_keep = ['/Annotation/@Name',
                   '/Annotation/#id', 
                   '/Annotation/Regions/Region/#id',
                  '/Annotation/Regions/Region/Vertices/V/@X',
                  '/Annotation/Regions/Region/Vertices/V/@Y']
        
        
        df = (anno.loc[:, columns_to_keep]
              .rename(columns={'/Annotation/@Name': 'Annotation',
                          '/Annotation/#id': 'Id',
                          '/Annotation/Regions/Region/#id': 'Region_ID',
                          '/Annotation/Regions/Region/Vertices/V/@X': 'X',
                          '/Annotation/Regions/Region/Vertices/V/@Y': 'Y'})
             )
                    

        df['Image_ID'] = str(newname)
    
        prf.append(df)
        
    d = pd.concat(prf)
    
    return d

    
data = loading_data()
d= loading_annotation()



```

### Function to Calculate Centroids of the rectangular bounding boxes
### Classification of Myeloid subtypes

```python
def X_centroid(row):
    X = int(row['XMin'] + row['XMax'])/2

    return int(X)

def Y_centroid(row):
    Y = int(row['YMin'] + row['YMax'])/2
    return int(Y)

### Function of myeloid cell Classification


def classification(row):
    
    if (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] == 0) & (row['Opal_620_Classification'] >=1) & (row['Opal_690_Classification'] == 0):
        return 'CM0'
    elif (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] == 0) & (row['Opal_620_Classification'] >=1) & (row['Opal_690_Classification'] >=1):
        return 'CM1'
    elif (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] >= 1) & (row['Opal_620_Classification'] ==0) & (row['Opal_690_Classification'] ==0):
        return 'CM4'
    elif (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] == 0) & (row['Opal_620_Classification'] ==0) & (row['Opal_690_Classification'] == 0):
        return 'CM_other'
    elif (row['Opal_520_Classification'] ==0) & (row['Opal_570_Classification'] >= 1) & (row['Opal_620_Classification'] ==0) & (row['Opal_690_Classification'] == 0):
        return 'RNASE'
    elif (row['Opal_520_Classification'] ==0) & (row['Opal_570_Classification'] == 0) & (row['Opal_620_Classification'] >=1) & (row['Opal_690_Classification'] == 0):
        return 'CD300'
    elif (row['Opal_520_Classification'] ==0) & (row['Opal_570_Classification'] == 0) & (row['Opal_620_Classification'] == 0) & (row['Opal_690_Classification'] >=1):
        return 'CD36'
    elif (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] == 0) & (row['Opal_620_Classification'] == 0) & (row['Opal_690_Classification'] >=1):
        return 'CM_CD36'
    elif (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] >=1) & (row['Opal_620_Classification'] >=1) & (row['Opal_690_Classification'] ==0):
        return 'CM_RNASE_CD300'
    elif (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] >=1) & (row['Opal_620_Classification'] ==0) & (row['Opal_690_Classification'] >=1):
        return 'CM_RNASE_CD36'
    elif (row['Opal_520_Classification'] >=1) & (row['Opal_570_Classification'] >=1) & (row['Opal_620_Classification'] >=1) & (row['Opal_690_Classification'] >=1):
        return 'CM_RNASE_CD300_CD36'
    else:
        return 'NC'
    




def calculate_centroid_classification():
    
    prf = []

    for i, p in enumerate(data['Image_ID'].unique()):
        
        dt = data.loc[data['Image_ID'] == str(p)]
    
    # Use the height and width to calculate the area

        dt['X'] = dt.apply(X_centroid, axis=1)
        dt['Y'] = dt.apply(Y_centroid, axis=1)
        
        dt['Classification'] = dt.apply(classification, axis=1)
    
        prf.append(dt)

    data_new = pd.concat(prf)
      
    
    return data_new
    
data = calculate_centroid_classification() 

data.head()
```

## Creating Mask 

### Note:
1) Annotated regions are filled using polygon function(Skimage)

2) Cell coordinates were matched with masked region and separate dataframe is generated

#### Glomerulus Mask

```python
from skimage.draw import polygon


def Intersection(dtt, mask): 
    return set(dtt).intersection(set(mask))


fmask = []

gf =[]

for i, p in enumerate(d.Image_ID.unique()):
    anno = d.loc[d['Image_ID'] == str(p)]
    dt= data.loc[data['Image_ID'] == str(p)]
    
    
    glom_id = anno.loc[anno['Annotation'] == "gloms"]['Region_ID'].unique().tolist()
    
    if len(glom_id) > 0:
        
        for gl in glom_id:
                
        
            poly = anno[(anno.Annotation == 'gloms') & (anno.Region_ID == gl)]
            r, c = polygon(np.array(poly.X),
                         np.array(poly.Y))
            mask = [m for m in zip(r, c)]
    
            dtt = [t for t in zip(dt.X, dt.Y)]
            d_mask = Intersection(dtt,mask)
            g = pd.DataFrame(d_mask, columns=['X', 'Y'])
            g['mask'] = 'gloms'
            g['Region_ID'] = str(gl)
            g['Image_ID'] = str(p)
        
        gf.append(g)
        
gmask = pd.concat(gf)

gmask.head()
```

```python
gmask.Image_ID.unique()
```

#### Vessel Mask

```python
vf =[]

for i, p in enumerate(d.Image_ID.unique()):
    anno = d.loc[d['Image_ID'] == str(p)]
    dt= data.loc[data['Image_ID'] == str(p)]
    
    vessel_id = anno.loc[anno['Annotation'] == "vessels"]['Region_ID'].unique().tolist()
    
    if len(vessel_id) > 0:
        
        for ve in vessel_id:
            
            
            poly = anno[(anno.Annotation == 'vessels') & (anno.Region_ID == ve)]
            r, c = polygon(np.array(poly.X),
                         np.array(poly.Y))
            mask = [m for m in zip(r, c)]
    
            dtt = [t for t in zip(dt.X, dt.Y)]
            d_mask = Intersection(dtt,mask)
            v = pd.DataFrame(d_mask, columns=['X', 'Y'])
            v['mask'] = 'vessels'
            v['Region_ID'] = str(ve)
            v['Image_ID'] = str(p)
        
            vf.append(v)
           
        
vmask = pd.concat(vf)
vmask


fmask = pd.DataFrame(gmask.append(vmask)) 

fmask['mask'].unique()

fmask.head()
```

## Calculating Distances Glomerulus

```python
import math

glom_id = sorted(final.loc[final['mask'] == "gloms"]['Region_ID'].unique().tolist())

prf = []

for p in glom_id:
    df = final.loc[(final['mask'] == 'gloms') & (final['Region_ID'] == p)]
    Classification = []
    for i in df['Classification']:
        for j in df['Classification']:
            cl = [i, j]
            Classification.append(cl)
            
    mask = []
    for i in df['mask']:
        for j in df['mask']:
            m = [i, j]
            mask.append(m)
            
    Region_ID = []
    for i in df['Region_ID']:
        for j in df['Region_ID']:
            r = [i, j]
            Region_ID.append(r)
            
    X = list(df.X)       
    Y = list(df.Y)

    distance = []  
    for i in range(0, len(X)):
        for j in range(0, len(X)):
            dist = math.sqrt((X[j] - X[i])**2 + ((Y[j] - Y[i])**2)) * 0.495952562137432
            distance.append(dist)

       
    coordinates = []        
    for i in zip(df.X, df.Y):
        for j in zip(df.X, df.Y):
            cor = [i, j]
            coordinates.append(cor)
        
    obj = []        
    for i in range(0, len(df.Object_Id)):
        for j in range(0, len(df.Object_Id)):
            com_obj = [i, j]
            obj.append(com_obj)
        
    obj_id = pd.DataFrame(obj, columns=['objID_1', 'objID_2'])
    dist = pd.DataFrame(distance, columns=['Distance'])
    meta = pd.DataFrame(Classification, columns=['idx_1', 'idx_2'])
    cord = pd.DataFrame(coordinates, columns=['cord_1', 'cord_2'])
    mask = pd.DataFrame(mask, columns=['mask_1', 'mask_2'])
    RID = pd.DataFrame(Region_ID, columns=['RID_1', 'RID_2'])

    pf = pd.concat([obj_id, cord, meta,mask, RID, dist], axis =1)
    
    prf.append(pf)
    
    
prf = pd.concat(prf)     
```

```python

prf[['X1', 'Y1']] = pd.DataFrame(prf['cord_1'].tolist(), index=prf.index) 
prf[['X2', 'Y2']] = pd.DataFrame(prf['cord_2'].tolist(), index=prf.index) 
```

```python
import math

fout = final.loc[(final['mask'] != "gloms") &(final['Classification'] != "NC")]

out_id = sorted(fout.loc[final['mask'] != "gloms"]['mask'].unique().tolist())
out_id 
orf = []

for p in out_id:
    #df = fout.loc[(fout['mask'] != 'gloms') & (fout['Region_ID'] == p)]
    df = fout.loc[(fout['mask'] == str(p))]
    Classification = []
    for i in df['Classification']:
        for j in df['Classification']:
            cl = [i, j]
            Classification.append(cl)
            
    mask = []
    for i in df['mask']:
        for j in df['mask']:
            m = [i, j]
            mask.append(m)
            
#     Region_ID = []
#     for i in df['Region_ID']:
#         for j in df['Region_ID']:
#             r = [i, j]
#             Region_ID.append(r)
            
    X = list(df.X)       
    Y = list(df.Y)

    distance = []  
    for i in range(0, len(X)):
        for j in range(0, len(X)):
            dist = math.sqrt((X[j] - X[i])**2 + ((Y[j] - Y[i])**2)) * 0.495952562137432
            distance.append(dist)

       
    coordinates = []        
    for i in zip(df.X, df.Y):
        for j in zip(df.X, df.Y):
            cor = [i, j]
            coordinates.append(cor)
        
    obj = []        
    for i in range(0, len(df.Object_Id)):
        for j in range(0, len(df.Object_Id)):
            com_obj = [i, j]
            obj.append(com_obj)
        
    obj_id = pd.DataFrame(obj, columns=['objID_1', 'objID_2'])
    dist = pd.DataFrame(distance, columns=['Distance'])
    meta = pd.DataFrame(Classification, columns=['idx_1', 'idx_2'])
    cord = pd.DataFrame(coordinates, columns=['cord_1', 'cord_2'])
    mask = pd.DataFrame(mask, columns=['mask_1', 'mask_2'])
    #RID = pd.DataFrame(Region_ID, columns=['RID_1', 'RID_2'])

    of = pd.concat([obj_id, cord, meta,mask,dist], axis =1)
    
    orf.append(of)
    
    
orf = pd.concat(orf)  

```
