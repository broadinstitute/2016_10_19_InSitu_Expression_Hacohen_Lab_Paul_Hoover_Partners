```python
import numpy as np
import os
import pandas as pd
import numpy as np
import glob
import plotnine as gg
import re
from plotnine import *
import fnmatch
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors.kde import KernelDensity
```

### Paths of Data and Annotations

```python
def fileslist():
    base = '../data/ExpPHa07_1/'
    datacsvs = os.listdir('../data/ExpPHa07_1/')
    flist = []
    for f in datacsvs:
        if f.endswith('csv'):
            fpath = os.path.join(base, f)
            flist.append(fpath)
    return flist

def annolist():
    base = '../data/ExpPHa07_1/annotations/'
    annocsvs = os.listdir('../data/ExpPHa07_1/annotations/')
    annolist = []
    for i in annocsvs:
        if not "job" in i:
            annopath = os.path.join(base, i)
            annolist.append(annopath)
    return annolist

flist = fileslist()
annolist = annolist()

fpath = flist[0]
annopath = annolist[3]
print(annopath)
print(fpath)
```

### Loading Tissue Data

```python
data = pd.read_csv(fpath)

# Renaming of column names

data.columns = [col.replace(' ', '_') for col in data.columns]

## Extracting tissue image names from filenames

data['Image_Location'] = pd.DataFrame(data.loc[:, 'Image_Location']
                                      .str.split('\\')
                                      .str.get(9)
                                      .str.replace('.tif', ''))

data = (data
        .rename(columns={'Image_Location': 'Image_ID'})
        .drop(columns = ['Analysis_Region', 'Analysis_Inputs'])  
       )

data.head()
```

### Annotations csv file

### Function to Calculate Centroids of the rectangular bounding boxes

```python
# Use the height and width to calculate the area
def X_centroid(row):
    X = int(row['XMin'] + row['XMax'])/2

    return int(X)

def Y_centroid(row):
    Y = int(row['YMin'] + row['YMax'])/2
    return int(Y)

data['X'] = data.apply(X_centroid, axis=1)
data['Y'] = data.apply(Y_centroid, axis=1)

```

```python
def classification_myleoid(row): 
    if (row['Opal_520_Classification'] > 2):
        return 'all'
    else:
        return 'NC'
def classification_subtypes(row):
    
    if (row['Opal_520_Classification'] > 2) & (row['Opal_570_Classification'] < 2) & (row['Opal_620_Classification'] > 2) & (row['Opal_690_Classification'] < 2):
        return 'CM0'
    elif (row['Opal_520_Classification'] > 2) & (row['Opal_570_Classification'] < 2) & (row['Opal_620_Classification'] > 2) & (row['Opal_690_Classification'] > 2):
        return 'CM1'
    elif (row['Opal_520_Classification'] > 2) & (row['Opal_570_Classification'] > 2) & (row['Opal_620_Classification'] < 2) & (row['Opal_690_Classification'] < 2):
        return 'CM4'
    elif (row['Opal_520_Classification'] > 2) & (row['Opal_570_Classification'] < 2) & (row['Opal_620_Classification'] < 2) & (row['Opal_690_Classification'] < 2):
        return 'others'
    else:
        return 'NC'
    
def threesubtypesonly(row):
    
    if (row['Opal_520_Classification'] > 2) & (row['Opal_570_Classification'] < 2) & (row['Opal_620_Classification'] > 2) & (row['Opal_690_Classification'] < 2):
        return 'CM0'
    elif (row['Opal_520_Classification'] > 2) & (row['Opal_570_Classification'] < 2) & (row['Opal_620_Classification'] > 2) & (row['Opal_690_Classification'] > 2):
        return 'CM1'
    elif (row['Opal_520_Classification'] > 2) & (row['Opal_570_Classification'] > 2) & (row['Opal_620_Classification'] < 2) & (row['Opal_690_Classification'] < 2):
        return 'CM4'
    else:
        return 'NC+others'

    
### Applying Classification Function
    
data['Classification_1'] = data.apply(classification_myleoid, axis=1)
data['Classification_2'] = data.apply(classification_subtypes, axis=1)
data['Classification_3'] = data.apply(threesubtypesonly, axis=1)

```

## Creating Mask

```python
from skimage.draw import polygon


def Intersection(data, mask): 
    return set(dt).intersection(set(mask))

fmask = []

## Glomerulus Mask

glom_id = d.loc[d['Annotation'] == "gloms"]['Region_ID'].unique().tolist()


gf = []
for p in glom_id:
    poly = d[(d.Annotation == 'gloms') & (d.Region_ID == p)]
    r, c = polygon(np.array(poly.X),
                         np.array(poly.Y))
    mask = [m for m in zip(r, c)]
    dt = [t for t in zip(data.X, data.Y)]
    d_mask = Intersection(dt,mask)
    g = pd.DataFrame(d_mask, columns=['X', 'Y'])
    g['mask'] = 'gloms'
    g['Region_ID'] = p
        
    gf.append(g)
gmask = pd.concat(gf)


# Vessels Mask

vessel_id = d.loc[d['Annotation'] == "vessels"]['Region_ID'].unique().tolist()
vessel_label = ['vessels'] * len(vessel_id)
vf = []

for p in vessel_id:
   
    poly = d[(d.Annotation == 'vessels') & (d.Region_ID == p)]
    row, col = polygon(np.array(poly.X),
                         np.array(poly.Y))
    mask = [m for m in zip(row, col)]
    dt = [t for t in zip(data.X, data.Y)]
    d_mask = Intersection(dt,mask)
    v = pd.DataFrame(d_mask, columns=['X', 'Y'])
    v['mask'] = 'vessels'
    v['Region_ID'] = p
        
    vf.append(v)
vmask= pd.concat(vf)


fmask = pd.DataFrame(gmask.append(vmask))    
```

## Merging of DataFrames

```python

final =  data.merge(fmask,how='left', on=['X', 'Y'])
(final['mask']
 .fillna('Out', inplace=True)
)
(final['Region_ID']
 .fillna(0, inplace=True)
)

final['Region_ID'] = final['Region_ID'].astype(int)
```

### Making Extracolumns for visualization Purposes

```python
def masking_myeloid_tissue(row):
    if (row['Classification_1'] == "all"):
        return "all"
    else:
        return None
    
def masking_myeloid_glom(row):
    if (row['Classification_1'] == "all") & (row['mask'] == "gloms"):
        return "all"
    else:
        return None
    
def masking_CM0_glom(row):
    if (row['Classification_2'] == "CM0") & (row['mask'] == "gloms"):
        return "CM0"
    else:
        return None
def masking_CM1_glom(row):
    if (row['Classification_2'] == "CM1") & (row['mask'] == "gloms"):
        return "CM1"
    else:
        return None
def masking_CM4_glom(row):
    if (row['Classification_2'] == "CM4") & (row['mask'] == "gloms"):
        return "CM4"
    else:
        return None
    


final['Myleoid_glom'] = final.apply(masking_myeloid_glom, axis=1)
final['Myleoid_tissue'] = final.apply(masking_myeloid_tissue, axis=1)
final['CM0_glom'] = final.apply(masking_CM0_glom, axis=1)
final['CM1_glom'] = final.apply(masking_CM1_glom, axis=1)
final['CM4_glom'] = final.apply(masking_CM4_glom, axis=1)
```

```python
def out_anno(row):
    if row['Annotation'] == "outline":
        return "outline"
    else:
        return None
def glom_anno(row):
    if row['Annotation'] == "outline":
        return "outline"
    elif row['Annotation'] == "gloms":
        return "Glomerulus"
    else:
        return None
    
d['out'] = d.apply(out_anno, axis=1)
d['glom'] = d.apply(glom_anno, axis=1) 
```

```python
def mask2(row):
    if row['mask'] != "gloms":
        return "Extra-Glomerulus"
    else:
        return "Glomerulus"
final['mask2'] = final.apply(mask2, axis=1)
final['mask2'].unique()
```

```python
import numpy as np
import scipy.spatial as spatial

scatter_x = np.array(dm.X)
scatter_y = np.array(dm.Y)

points = np.array(list(zip(scatter_x, scatter_y)))
tree = spatial.KDTree(np.array(points))
radius = 25

neighbors = tree.query_ball_tree(tree, radius)


frequency = np.array(list(map(len, neighbors)))

density = frequency/(radius**2)*3.14
dn = pd.DataFrame(density, columns = ['Density'])
dm_new = pd.merge(dm, dn, left_index=True, right_index=True)
dm_new
#dm_new = dm.join(dn)
```

```python

estimators = []

for c in dm['Classification_3'].unique():
    temp = dm[dm['Classification_3'] == str(c)]
    temp1 = temp[['X', 'Y']].to_numpy()
    kd = KernelDensity(kernel='gaussian', bandwidth=5).fit(temp1)
    estimators.append(kd)

scr = []   
for estimator in estimators:
    x = estimator.score_samples(temp1)
    scr.append(x)   
scores = pd.DataFrame(scr)    


def density_score(scores):
    scores.rename(index={0: "NC+others", 1:"CM0", 2:"CM1", 3: "CM4"}, inplace=True)
    sc = scores.sum(axis=1)
    NC = sc.loc['NC+others']
    CM0 = sc.loc['CM0']
    CM1 = sc.loc['CM1']
    CM4 = sc.loc['CM4']
    
    return  sc #NC, others, CM0, all, CM1, CM4
    
    
p = density_score(scores)  

pnew = pd.DataFrame(p, columns=['Density'])
p1=pnew.reset_index()
p1
```

## Density Function for each Cell Subtypes in Glomerulus

```python
from statistics import mean 

dg = final[(final['mask'] == 'gloms')]


scores = []
for i, p in enumerate(dg['Region_ID'].unique()):
    dgn = dg.loc[dg['Region_ID'] == p]
      
    for c in dgn['Classification_3'].unique():
        
        temp = dgn[dgn['Classification_3'] == str(c)]
        temp1 = temp[['X', 'Y']].to_numpy()
        kde = KernelDensity(kernel='gaussian', bandwidth=0.1).fit(temp1)
        logprobX = kde.score_samples(temp1)
        probX = np.exp(logprobX)
        gx = pd.DataFrame(probX, columns=['Density'])
        gx['Cell_Subtypes'] = c
        gx['Region_ID'] = p
        scores.append(gx)
        
scores = pd.concat(scores)
     
        
for i, p in enumerate (scores['Region_ID'].unique()):
    uu = scores[scores['Region_ID'] == p]
    fig, ax = plt.subplots()
    sns.barplot(x='Cell_Subtypes',y='Density', data=uu, palette='Spectral')
    plt.xlabel('Cell Subtypes')
    plt.ylabel('Densities of Myleoid Cells subtypes \n in Kidney tissue')
    plt.title(str(p))
    plt.savefig(os.path.join(figpath, 'Cellsubtypes Densities {0}.png'.format(i)),bbox_inches = 'tight')
  
    

# for p in scores['Cell_Subtypes'].unique():
#     new = scores[scores['Cell_Subtypes'] == str(p)]
#     dnew = mean(new['Density'])
#     lst.append(dnew)
    
# celltypes = pd.DataFrame(scores['Cell_Subtypes'].unique().tolist(), columns=['Cell types'])
# new = pd.DataFrame(lst, columns=['Average'])

# newdensity = pd.merge(celltypes, new, left_index=True, right_index=True)
# newdensity        
           

# sns.barplot(x='Cell types',y='Average', data=newdensity, palette='Set2')
# plt.xlabel('Cell Subtypes')
# plt.ylabel('Mean Densities')
# plt.title("Mean Densities of Cell subtypes")

```

### Density of Glomerulus

```python
from statistics import mean 

dg = final[(final['mask'] == 'gloms')]


scores = []
for i, p in enumerate(dg['Region_ID'].unique()):
    
    dgn = dg.loc[dg['Region_ID'] == p]
    temp1 = dgn[['X', 'Y']].to_numpy()
    kde = KernelDensity(kernel='gaussian', bandwidth=0.1).fit(temp1)
    logprobX = kde.score_samples(temp1)
    probX = np.exp(logprobX)
    gx = pd.DataFrame(probX, columns=['Density'])
    gx['Region_ID'] = p
    scores.append(gx)
        
scores = pd.concat(scores)
  

glom_density= scores.groupby(['Region_ID']).mean().reset_index()
glom_density
sns.barplot(x='Region_ID',y='Density', data=glom_density, palette='Set2')
plt.xlabel('Region_ID')
plt.ylabel('Mean Densities')
plt.title("Mean Densities of Glomeruli")
plt.savefig(os.path.join(figpath, 'Mean_Densities_Glomerulus.png'))
```
