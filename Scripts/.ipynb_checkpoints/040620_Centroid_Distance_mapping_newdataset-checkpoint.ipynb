### Function to Calculate Centroids of the rectangular bounding boxes

```python
# Use the height and width to calculate the area
def X_centroid(row):
    X = int(row['XMin'] + row['XMax'])/2

    return int(X)

def Y_centroid(row):
    Y = int(row['YMin'] + row['YMax'])/2
    return int(Y)

data['X'] = data.apply(X_centroid, axis=1)
data['Y'] = data.apply(Y_centroid, axis=1)

```

### Classification of Myeloid cells

```python

data['Classification_1'] = data.apply(lambda x: 
                                      "all"if x.Opal_520_Classification>2 
                                      else "NC",axis=1)



def classification_subtypes(row):
    
    if (row['Opal_520_Classification'] > 2) & (row['Opal_570_Classification'] < 2) & (row['Opal_620_Classification'] > 1) & (row['Opal_690_Classification'] < 2):
        return 'CM0'
    elif (row['Opal_520_Classification'] > 2) & (row['Opal_570_Classification'] < 2) & (row['Opal_620_Classification'] > 1) & (row['Opal_690_Classification'] > 1):
        return 'CM1'
    elif (row['Opal_520_Classification'] > 2) & (row['Opal_570_Classification'] > 1) & (row['Opal_620_Classification'] < 2) & (row['Opal_690_Classification'] < 2):
        return 'CM4'
    elif (row['Opal_520_Classification'] > 2) & (row['Opal_570_Classification'] < 2) & (row['Opal_620_Classification'] < 2) & (row['Opal_690_Classification'] < 2):
        return 'myeloid_others'
    else:
        return 'NC'
    
def threesubtypesonly(row):
    
    if (row['Opal_520_Classification'] > 2) & (row['Opal_570_Classification'] < 2) & (row['Opal_620_Classification'] > 1) & (row['Opal_690_Classification'] < 2):
        return 'CM0'
    elif (row['Opal_520_Classification'] > 2) & (row['Opal_570_Classification'] < 2) & (row['Opal_620_Classification'] > 1) & (row['Opal_690_Classification'] > 1):
        return 'CM1'
    elif (row['Opal_520_Classification'] > 2) & (row['Opal_570_Classification'] > 1) & (row['Opal_620_Classification'] < 2) & (row['Opal_690_Classification'] < 2):
        return 'CM4'
    else:
        return 'NC+others'

    
# ### Applying Classification Function
    
data['Classification_2'] = data.apply(classification_subtypes, axis=1)
data['Classification_3'] = data.apply(threesubtypesonly, axis=1)



```

### Creating Mask

```python
from skimage.draw import polygon
from scipy.spatial.distance import cdist
import math
import matplotlib.pyplot as plt
import pandas as pd



def Intersection(data, mask): 
    return set(dt).intersection(set(mask))

def distance(row):
    dist = math.sqrt((row['X']- row['X_centroid'])**2 + ((row['Y'] - row['Y_centroid'])**2)) * 0.495952562137432
    return dist
```

## Glomerulus Mask & Centroid Distance Mapping

```python

glom_id = d.loc[d['Annotation'] == "gloms"]['Region_ID'].unique().tolist()

gf = []


for i, p in enumerate(glom_id):
    anno = d[(d.Annotation == 'gloms') & (d.Region_ID == p)]
    anno['mask'] = "outline"
#     anno['Region_ID'] = p
    r, c = polygon(np.array(anno.X), np.array(anno.Y))
    mask = [m for m in zip(r, c)]
    dt = [t for t in zip(data.X, data.Y)]
    d_mask = Intersection(dt,mask)
    dtt = pd.DataFrame(d_mask, columns = ['X', 'Y'])
    dtt['mask'] = "gloms"
    dtt['Region_ID'] = p
    db1 = anno.loc[:, 'X':]
    db2 = dtt.loc[:, 'X':]
    db = db1.append(db2)
    
    ## Creating centroids of Glomerulus


    db['X_centroid'] = db.apply(lambda x: int(np.mean(db.X)), axis=1)
    db['Y_centroid'] = db.apply(lambda x: int(np.mean(db.Y)), axis=1)
    
    ## Calculating distance
    
    db['Distance'] = db.apply(distance, axis=1)
    
    ## Minimum Max Scaling
    
    db['Distance_scaled'] = db['Distance'].apply(lambda x: int(x -  np.min(db.Distance))/(np.max(db.Distance) - np.min(db.Distance)))
    
    ## Z-normalization
    
    db['Distance_normalized'] = db['Distance'].apply(lambda x: int(x -  np.mean(db.Distance))/np.std(db.Distance))
    
    
    
    dm = pd.DataFrame(db, columns = ['X', 'Y', 'mask', 'Region_ID', 'X_centroid', 'Y_centroid', 'Distance', 'Distance_scaled', 'Distance_normalized'])
    
    
    gf.append(dm)
    
    
gmask = pd.concat(gf)







```

## Vessels Mask & Centroid Distance Mapping

```python
vessel_id = d.loc[d['Annotation'] == "vessels"]['Region_ID'].unique().tolist()

vf = []


for i, p in enumerate(vessel_id):
    anno = d[(d.Annotation == 'vessels') & (d.Region_ID == p)]
    anno['mask'] = "outline"
#     anno['Region_ID'] = p
    r, c = polygon(np.array(anno.X), np.array(anno.Y))
    mask = [m for m in zip(r, c)]
    dt = [t for t in zip(data.X, data.Y)]
    d_mask = Intersection(dt,mask)
    dtt = pd.DataFrame(d_mask, columns = ['X', 'Y'])
    dtt['mask'] = "vessels"
    dtt['Region_ID'] = p
    db1 = anno.loc[:, 'X':]
    db2 = dtt.loc[:, 'X':]
    db = db1.append(db2)
    
    ## Creating centroids of Glomerulus


    db['X_centroid'] = db.apply(lambda x: int(np.mean(db.X)), axis=1)
    db['Y_centroid'] = db.apply(lambda x: int(np.mean(db.Y)), axis=1)
    
    ## Calculating distance
    
    db['Distance'] = db.apply(distance, axis=1)
    
    ## Minimum Max Scaling
    
    db['Distance_scaled'] = db['Distance'].apply(lambda x: int(x -  np.min(db.Distance))/(np.max(db.Distance) - np.min(db.Distance)))
    
    ## Z-normalization
    
    db['Distance_normalized'] = db['Distance'].apply(lambda x: int(x -  np.mean(db.Distance))/np.std(db.Distance))
    
    
    
    dm = pd.DataFrame(db, columns = ['X', 'Y', 'mask', 'Region_ID', 'X_centroid', 'Y_centroid', 'Distance', 'Distance_scaled', 'Distance_normalized'])
    
    
    vf.append(dm)
    
vmask = pd.concat(vf)


fmask = pd.DataFrame(gmask.append(vmask))  

## Excluding the outline coordinates

prf = fmask.query('mask != "outline"')

    
```
